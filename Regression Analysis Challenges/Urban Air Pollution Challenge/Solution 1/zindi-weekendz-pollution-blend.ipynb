```python
import numpy as np
import pandas as pd
import os
```

```python
ID_COL = 'Place_ID X Date'
```

```python
df = pd.read_csv('preds_lgbm_v1.csv')[[ID_COL]]
df['v1'] = pd.read_csv('preds_lgbm_v1.csv')['target']
df['v2'] = pd.read_csv('preds_lgbm_v2.csv')['target']
df['v3'] = pd.read_csv('preds_lgbm_v3.csv')['target']
df['v4'] = pd.read_csv('preds_lgbm_v4.csv')['target']
```

# Blending the different versions

## The weights were decided based on the performance of each version on the leaderboard

```python
sub_df = df[[ID_COL]]
sub_df['target'] = ((df['v1']*0.75 + df['v2']*0.25)*0.6 + df['v4']*0.4)*0.9 + df['v3']*0.1
```

```python
sub_df.to_csv('submission.csv', index=False)
```
